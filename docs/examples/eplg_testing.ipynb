```python
from iqm.benchmarks.utils import get_iqm_backend
from iqm.benchmarks.randomized_benchmarking.direct_rb.direct_rb import generate_fixed_depth_parallel_drb_circuits
```

```python
qubits_arr = [[0,1], [8,9], [17,18]]
num_pairs = len(qubits_arr)
backend = get_iqm_backend("fakeapollo")
density_2q = {str(qubits_arr[0]): 0.15, str(qubits_arr[1]): 0.5, str(qubits_arr[2]): 0.5}
two_qg = {str(q): {"CZGate": 1.0} for q in qubits_arr}
clifford_1q_prob = {str(q): 0.0 for q in qubits_arr}
sqg_ensemble={str(q): {"HGate": 1.0} for q in qubits_arr}
```

```python
(drb_circs, drb_circs_t), _ = generate_fixed_depth_parallel_drb_circuits(qubits_arr, 10, 5, backend, density_2q, two_qg, clifford_1q_prob, sqg_ensemble, qiskit_optim_level=3)
```
